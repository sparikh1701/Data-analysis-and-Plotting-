```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,1,21) 
y = 2*x
np.random.seed(seed = 1)
noise = np.random.normal(0, 0.1, x.size)
z = y + noise
print(noise)
print(z)
print(y)
x
```

```python
plt.subplot(1,1,1)
plt.plot(x,color = 'g', label = "X values")
plt.plot(y, color = 'b', label = " y values")
plt.legend()
plt.xlabel('Data point #')
plt.ylabel('Data point value')
plt.title('Data points of x and y with associated label')
```

```python
def linr(x,A,B):
    return A * x + B
A = 2
B = 4
plt.plot(1,1,1)
lin_1,_=curve_fit(linr,y,x)
y_expected = linr(y,lin_1[0],lin_1[1])
plt.scatter(y,x, label = 'Y data', color = 'b')
plt.plot(y,y_expected,color = 'r')
lin_2,tums=curve_fit(linr,z,x)
z_expected = linr(z,lin_2[0],lin_2[1])
plt.scatter(z,x, label = 'z data', color = 'g')
plt.plot(z,z_expected,color = 'y')
plt.legend()
plt.xlabel('Data point #')
plt.ylabel('Data point value')
plt.title('Data points of pure data and noisy data with associated label')
```

```python
plt.subplot(1,1,1)
plt.plot(y,color='b', label = 'pure data')
plt.plot(z,color = 'r', label = 'noisy data')
plt.legend()
plt.xlabel('Data point #')
plt.ylabel('Data point value')
plt.title('Data points of pure data and noisy data with associated label')
```

```python
def quad(x,A,B,C,D):
    return (A * (x**4))+ (B*(x**3)) + (C*(x**2)) - x + D
```

```python
np.random.seed(seed = 1)
noise_new_2 = np.random.normal(0, .01, x.size)
z_new_2 = y + noise_new_2
```

```python
plt.subplot(1,2,1)
lin_1,Cam=curve_fit(linr,y,x)
y_expected = linr(y,lin_1[0],lin_1[1])
plt.scatter(y,x, label = 'Y data linear', color = 'purple')
plt.plot(y,y_expected,color = 'green')
lin_2,tums=curve_fit(linr,z,x)
z_expected = linr(z,lin_2[0],lin_2[1])
plt.scatter(z,x, label = 'z data linear', color = 'brown')
plt.plot(z,z_expected,color = 'black')
plt.legend()
# plt.xlabel('Data point #')
# plt.ylabel('Data point value')
# plt.title('Data points of pure data and noisy data with associated label')

plt.subplot(1,2,2)
lin_5,olin=curve_fit(quad,y,x)
y_expected_quad = quad(y,lin_5[0],lin_5[1],lin_5[2],lin_5[3])
plt.scatter(y,x, label = 'Y data quad', color = 'blue')
plt.plot(y,y_expected_quad,color = 'pink')
lin_6,yum=curve_fit(quad,z_new_2,x)
z_expected_3 = quad(z_new_2,lin_6[0],lin_6[1],lin_6[2],lin_6[3])
plt.scatter(z_new_2,x, label = 'z data quad', color = 'red')
plt.plot(z_new_2,z_expected_3,color = 'orange')
plt.legend()
plt.xlabel('Data point #')
plt.ylabel('Data point value')
plt.title('Data points of pure data and noisy data with associated label')
```
